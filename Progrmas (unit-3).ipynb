```python
# 3. WAP to chexk for 3 digit number only
#      1.palindrome
#      2.Armstong number
#      3.prime
#      4.Happy number

print("1.palindrome \n2.Armstong number \n3.prime \n4.Happy number\n")
c=int(input("Enter choice: "))
 
def palindrome(n):
    t=n
    rev_num=0
    while n!=0:
        digit=n%10
        rev_num=rev_num*10+digit
        n=n//10

    if rev_num==t:
        print("It is a Palindrome number")
    else:
        print("It is not a Palindrome number")   

def armstrong(i):
    t1=i
    sum=0
    while i!=0:
        digit=i%10
        sum=sum+digit**3
        i=i//10

    if sum==t1:
         print("It is a Armstrong number")
    else:
        print("It is not a Armstrong number")    
        
def prime(x):
    count=0
    for i in range(1,x+1):
        if x%i==0:
            count=count+1
    if count==2:
        print("Prime")
    else:
        print("Not prime")
        
def happy_number(x):
    sum=0
    while x!=1:
        while x!=0:
            i=x%10
            sum=sum+i**2
            x=x//10
        x=sum
        sum=0
        if(x==4):
            break

    if x==1:
        print("its a Happy number")
    else:
        print("its not a Happy number")

if c==1:
    palindrome(int(input("Enter n: ")))
elif c==2:
    armstrong(int(input("Enter n: ")))
elif c==3:
    prime(int(input("Enter n: ")))
elif c==4:
    happy_number(int(input("Enter n: ")))
else:
    print("Enter valid choice")
```

```python
# 6. Gross pay,Annual Income ,Income TaxCalculator Gross Pay,HRA,DA,other Allowance,
#    Profession Tax,providend fund,Transport allowance.
#  ---> Gross pay=Basic Pay+HRA+ DA +other allowance+ Transport Allowance -Profession Tax- Providend Fund
#  ---> Profession tax --->200Rs.for each grade level amd each month
#  ---> HRA ----> 0.3 times of each grade level basic pay ---> class 1 cities
#                 0.2 times of each grade level basic pay ---> class 2 cities
#                 0.1 times of each grade level basic pay ---> class 3 cities
#  ---> DA --> 0.5 times of basic pay---> for each grade levels 
#  ---> other allowances as per table
#  --->Providend fund ---> 0.11 times of basic pay for each grade 
#  ---> Transport Allowance ---> 900Rd for each level

# ----------------------------------------------------------------------
# Grade level              Basic pay                  other allowances
# ----------------------------------------------------------------------
#     A                    60000                       8000
#     B                    50000                       7000 
#     C                    40000                       6000   
#     D                    30000                       5000   
#     E                    20000                       4000
#     F                    10000                       3000  
# ----------------------------------------------------------------------

#                            Income Tax
# ----------------------------------------------------------------------
#     Amount                                     Income Tax Rate
# ----------------------------------------------------------------------
#    Upto Rs 2,50,000                                 0%
#    Upto Rs 2,50,000                                 5% above
#    Upto Rs 2,50,000                                 10%
#    Upto Rs 2,50,000                                 15%
#    Upto Rs 2,50,000                                 20%
#    Upto Rs 2,50,000                                 25%
#    Upto Rs 2,50,000                                 30%


professionaltax=200
basicpay=0
otherallowances=0
hra=0
taxrate=0
da=0.5
grosspay=0
transportallowances=900
providendfund=0.11
level=input("Enter Grade level: ")
city=int(input("Enter city class: "))

if level=="A":
    basicpay=60000
    otherallowances=8000
elif level=="B":
    basicpay=50000
    otherallowances=7000
elif level=="C":
    basicpay=40000
    otherallowances=6000
elif level=="D":
    basicpay=30000
    otherallowances=5000
elif level=="E":
    basicpay=20000
    otherallowances=4000
elif level=="F":
    basicpay=10000
    otherallowances=3000
else:
    print("Enter valid grade level.")
    
if city==1:
    hra=0.3
elif city==2:
    hra=0.2
elif city==3:
    hra=0.1
else:
    print("Enter valid city class.")
    
grosspay=basicpay+(hra*basicpay)+(da*basicpay)+otherallowances+transportallowances-professionaltax-(providendfund*basicpay)
annualpay=grosspay*12
print("Gross Pay: ",grosspay)
print("Annual Pay: ",annualpay)

if annualpay<=250000:
    taxrate=0
elif annualpay in range(250001,500000+1):
    taxrate=0.05*(annualpay-250000)
elif annualpay in range(500001,750000+1):
    taxrate=0.1*(annualpay-500000)+12500
elif annualpay in range(750001,1000000+1):
    taxrate=0.15*(annualpay-750000)+37500
elif annualpay in range(1000001,1250000+1):
    taxrate=0.2*(annualpay-1000000)+75000
elif annualpay in range(1250001,1500000+1):
    taxrate=0.25*(annualpay-1250000)+125000
elif annualpay>1500001:
    taxrate=0.3*(annualpay-150000)+187500
else:
    print("invalid Entry")

print("Tax amount: ",taxrate)
```
